```python
# Initialize list
expenses = []

# Input loop
while True:
    print("\nAdd an Expense:")
    date = input("Enter date (DD-MM): ")
    category = input("Enter category (Food, Travel, etc.): ").lower()
    amount = float(input("Enter amount (â‚¹): "))

    expenses.append({"date": date, "category": category, "amount": amount})

    more = input("Do you want to add another expense? (yes/no): ")
    if more.lower() != 'yes':
        break

# Summarize
total = 0
by_category = {}

for expense in expenses:
    total += expense["amount"]
    cat = expense["category"]
    by_category[cat] = by_category.get(cat, 0) + expense["amount"]

# Print summary
print(f"\nðŸ’¸ Total Spent: â‚¹{total}")
print("\nðŸ“Š Category-wise Spending:")
for cat, amt in by_category.items():
    print(f"{cat.capitalize()}: â‚¹{amt}")

# Save to .txt file
with open("expense_summary.txt", "w") as file:
    file.write(f"Total Spent: â‚¹{total}\n")
    file.write("Category-wise Spending:\n")
    for cat, amt in by_category.items():
        file.write(f"{cat}: â‚¹{amt}\n")

print("\nâœ… Summary saved to 'expense_summary.txt'")
# Pie chart of category-wise spending
import matplotlib.pyplot as plt

# Extract data
labels = list(by_category.keys())
sizes = list(by_category.values())

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Spending Breakdown by Category")
plt.axis('equal')  # Makes the pie circle perfect
plt.show()

```
